# **Pojam segmentacije**

**Segmentacija** se odnosi na proces razdvajanja slike na više delova ili regiona, pri čemu se svaki deo smatra "segmentom" ili "objektom" unutar slike. Segmentacija ima nekoliko ciljeva. Jedan od njih je olakšavanje analize i razumevanja slika. Pomoću segmentacije, moguće je izolovati objekte od pozadine i fokusirati se na njihove karakteristike i osobine. Ovo može biti korisno u raznim aplikacijama kao što su prepoznavanje objekata, praćenje pokreta, detekcija anomalija i druge.

Segmentacija ima široku primenu u različitim oblastima. U medicini, segmentacija se koristi za identifikaciju i izdvajanje specifičnih struktura u slikama, poput tumora na rendgenskim snimcima ili krvnih sudova na slikama magnetne rezonance. U industriji, segmentacija može biti korisna za inspekciju kvaliteta proizvoda ili identifikaciju defekata na površinama. U autonomnoj vožnji, segmentacija se koristi za prepoznavanje i praćenje saobraćajnih znakova, pešaka i drugih vozila.

Segmentacija može biti izazovna zbog nekoliko faktora. Varijacije u osvetljenju, senčenju i teksturi mogu otežati jasnu separaciju objekata od pozadine. Osim toga, segmentacija može biti vremenski zahtevna i zahtevati velike skupove podataka za obuku. Pored toga, postoji i izazov postizanja preciznih rezultata kada se objekti preklapaju ili imaju slične karakteristike.

# **Metode segmentacije**
**Metode bazirane na pikselima**-Ove metode pridružuju svakom pikselu u slici određenu oznaku koja ga klasifikuje kao deo određenog segmenta. Ove metode često koriste informacije o boji, intenzitetu, teksturi ili drugim atributima piksela kako bi doneli odluku o pripadnosti segmentu. Jedan od tradicionalnih pristupa je algoritam za pretragu oznaka (eng. *region growing*), koji iterativno proširuje regione piksela na osnovu njihovih sličnosti. Međutim, sa razvojem dubokog učenja, moderni pristupi koriste neuronske mreže za segmentaciju piksela, kao što su U-Net, SegNet i Mask R-CNN.

**Metode bazirane na regionima**-Ove metode grupišu susedne piksele koji dele slične karakteristike u regione ili objekte. Ove metode se često oslanjaju na segmentaciju piksela kako bi identifikovale inicijalne regione, a zatim primenjuju različite algoritme grupisanja ili optimizacije kako bi spojile slične regione i formirale konačne segmente. Na primer, metoda k-means grupiše piksele na osnovu njihovih vrednosti boje u predefinisan broj grupa. Metode kao što su algoritam grabcut (eng. GrabCut) i superpiksel segmentacija (eng. superpixel segmentation) takođe pripadaju ovoj kategoriji.

**Metode bazirane na ivicama**-Ove metode se fokusiraju na identifikaciju ivica objekata u slici kao kriterijum segmentacije. One koriste tehnike detekcije ivica, kao što su Canny detekcija ivica, Sobel filteri ili Laplasov operator, kako bi pronašle diskontinuitete u intenzitetu ili teksturi i označile ih kao ivice objekata. Nakon pronalaženja ivica, može se primeniti postupak povezivanja ivica kako bi se formirali konačni segmenti.

#**Evaluacija i metrike**
Evaluacija segmentacije je ključni korak kako bismo procenili kvalitet rezultata modela ili algoritma segmentacije. Nekoliko metrika koje se često koriste za merenje uspeha segmentacije:

**Tačnost** (*Accuracy*): Tačnost je jednostavna metrika koja izražava procenat tačno segmentiranih piksela u odnosu na ukupan broj piksela u slici. Međutim, tačnost može biti obmanjujuća ako postoji neuravnoteženost klasa, tj. ako je broj piksela u objektima i pozadini značajno neravnotežan.   **Preciznost** (*Precision*): Preciznost meri koliko su tačno identifikovani pozitivni pikseli (segmenti). Preciznost je korisna metrika kada je važno minimizirati lažno pozitivne rezultate. Formula za preciznost je: preciznost = TP / (TP + FP), gde TP predstavlja broj tačno pozitivnih piksela (segmentiranih piksela) i FP predstavlja broj lažno pozitivnih piksela.                             
**Odziv** (*Recall*): Odziv, poznat i kao osetljivost ili stopa istinitih pozitiva, meri koliko je dobro model identifikovao sve pozitivne piksele (segmente). Odziv je koristan kada je važno minimizirati lažno negativne rezultate. Formula za odziv je: odziv = TP / (TP + FN), gde TP predstavlja broj tačno pozitivnih piksela i FN predstavlja broj lažno negativnih piksela.       
        **F-mera** (*F-measure*): F-mera je harmonijska sredina preciznosti i odziva i koristi se za kombinovanje informacija o tačnosti i potpunosti rezultata segmentacije. Visoka vrednost F-mere ukazuje na dobro balansiranu preciznost i odziv. Formula za F-meru je: F-mera = 2 * (preciznost * odziv) / (preciznost + odziv).                       
        **Indeks preklapanja (IoU)**: Indeks preklapanja, poznat i kao Jaccardov indeks, meri preklapanje između predikcije segmentacije i stvarnih oznaka. Formula za IoU je: IoU = TP / (TP + FP + FN), gde TP predstavlja broj tačno pozitivnih piksela, FP predstavlja broj lažno pozitivnih piksela i FN predstavlja broj lažno negativnih piksela. Visoka vrednost IoU ukazuje na dobro preklapanje između predikcije segmentacije i stvarnih oznaka.

```python
import tensorflow as tf
```

Funkcija Conv2D_Block implementira blok konvolucijskih slojeva u dubokoj neuronskoj mreži za obradu slika. Prvo se primjenjuje konvolucijski sloj, zatim normalizacija po grupama i aktivacijska funkcija ReLU zatim ponovno konvolucijski sloj, normalizacija po grupama i ponovno aktivacijska funkcija ReLU. Na kraju se vraća izlazni tenzor. Ovaj blok omogućava ekstrakciju složenijih karakteristika iz ulaznih slika.


```python
def Conv2D_Block(input_tensor , n_filters):
  x = tf.keras.layers.Conv2D(filters = n_filters , kernel_size = (3 , 3) , kernel_initializer = 'he_normal' , padding = 'same')(input_tensor)
  x = tf.keras.layers.BatchNormalization()(x)
  x = tf.keras.layers.Activation('relu')(x)

  x = tf.keras.layers.Conv2D(filters = n_filters , kernel_size = (3 , 3) , kernel_initializer = 'he_normal' , padding = 'same')(x)
  x = tf.keras.layers.BatchNormalization()(x)
  x = tf.keras.layers.Activation('relu')(x)

  return x
```

Funkcija U_Net implementira U-Net arhitekturu za segmentaciju slika. Ukupno, ova funkcija gradi U-Net arhitekturu koja ima simetričnu strukturu s encoderskim i decoderskim delom. Encoderski deo se sastoji od konvolucijskih slojeva i grupiranja koji ekstrahiraju karakteristike iz slike, dok decoderski deo koristi dekonvoluciju i spajanje kako bi obnovio izlaznu sliku s detaljnim informacijama o segmentaciji.

```python
def U_Net(img_tensor , n_filters = 16):
  conv1 = Conv2D_Block(img_tensor , n_filters * 1)
  pool1 = tf.keras.layers.MaxPooling2D((2 , 2))(conv1)
  pool1 = tf.keras.layers.Dropout(0.05)(pool1)

  conv2 = Conv2D_Block(pool1 , n_filters * 2)
  pool2 = tf.keras.layers.MaxPooling2D((2 , 2))(conv2)
  pool2 = tf.keras.layers.Dropout(0.05)(pool2)

  conv3 = Conv2D_Block(pool2 , n_filters * 4)
  pool3 = tf.keras.layers.MaxPooling2D((2 , 2))(conv3)
  pool3 = tf.keras.layers.Dropout(0.05)(pool3)

  conv4 = Conv2D_Block(pool3 , n_filters * 8)
  pool4 = tf.keras.layers.MaxPooling2D((2 , 2))(conv4)
  pool4 = tf.keras.layers.Dropout(0.05)(pool4)

  conv5 = Conv2D_Block(pool4 , n_filters * 16)

  pool6 = tf.keras.layers.Conv2DTranspose(n_filters * 8 , (3 , 3) , (2, 2) , padding = 'same')(conv5)
  pool6 = tf.keras.layers.concatenate([pool6 , conv4])
  pool6 = tf.keras.layers.Dropout(0.05)(pool6)
  conv6 = Conv2D_Block(pool6 , n_filters * 8)

  pool7 = tf.keras.layers.Conv2DTranspose(n_filters * 4 , (3 , 3) , (2 , 2) , padding = 'same')(conv6)
  pool7 = tf.keras.layers.concatenate([pool7 , conv3])
  pool7 = tf.keras.layers.Dropout(0.05)(pool7)
  conv7 = Conv2D_Block(pool7 , n_filters * 4)

  pool8 = tf.keras.layers.Conv2DTranspose(n_filters * 2 , (3 , 3) , (2 , 2) , padding = 'same')(conv7)
  pool8 = tf.keras.layers.concatenate([pool8 , conv2])
  pool8 = tf.keras.layers.Dropout(0.05)(pool8)
  conv8 = Conv2D_Block(pool8 , n_filters * 2)

  pool9 = tf.keras.layers.Conv2DTranspose(n_filters * 1 , (3 , 3) , (2 , 2) , padding = 'same')(conv8)
  pool9 = tf.keras.layers.concatenate([pool9 , conv1])
  pool9 = tf.keras.layers.Dropout(0.05)(pool9)
  conv9 = Conv2D_Block(pool9 , n_filters * 1)

  output = tf.keras.layers.Conv2D(1 , (1 , 1) , activation = 'sigmoid')(conv9)

  u_net = tf.keras.Model(inputs = [img_tensor] , outputs = [output])

  return u_net
```

```python
import os
import re
```

 Metoda os.listdir() se koristi kako bi se dobila lista svih datoteka/fajlova (ili direktorijuma) unutar određenog direktorijuma.

```python
train_path = os.listdir('/content/train/')
test_path = os.listdir('/content/test/')
#ispis ukupnog broja fajlova pronadjenih unutar direktorijuma
print(len(train_path),len(test_path))
```

Sortiranje liste train_path na osnovu brojeva u nazivima fajlova, koristeći kompilirane regularne izraze i različite funkcije za obradu i sortiranje podataka

```python
reg = re.compile("[0-9]+")

temp1 = list(map(lambda x: reg.match(x).group(), train_path))
temp1 = list(map(int, temp1))

temp2 = list(map(lambda x: reg.match(x.split("_")[1]).group(), train_path))
temp2 = list(map(int, temp2))

train_path = [x for _,_,x in sorted(zip(temp1, temp2, train_path))]
train_path[:200]
```

```python
len(train_path)
```

Podela liste train_path na dve nove gde će se u jednoj nalaziti slike a u drugoj maske slika.

```python
train_image = []
train_mask = []

for index , item in enumerate(train_path):
  if index % 2 == 0:
    train_image.append(item)
  else:
    train_mask.append(item)
```

```python
print(len(train_image))
print(len(train_mask))
```

```python
import matplotlib.pyplot as plt
import random
from PIL import Image
import numpy as np
import skimage
```

Prikaz nasumično odabranih ulaznih slika i njihovih odgovarajućih maski.Cilj je dobiti uvid u karakteristike slika i njihovih maski kako bi se bolje razumela priroda podataka.

```python
fig , ax = plt.subplots(3 , 2 , figsize=[10 , 10])
fig.suptitle('Vizualizing Nerve Data')
ax[0 , 0].set_title('Train Data')
ax[0 , 1].set_title('Mask')

# print(ax.shape)
for i in range(0 , 3):
  ind = random.randint(0 , len(train_image))
  img1 = np.array(Image.open("/content/train/" + str(train_image[ind])))
  ax[i , 0].imshow(img1 , cmap = 'gray')

  img2 = np.array(Image.open("/content/train/" + str(train_mask[ind])))
  img2 = np.ma.masked_where(img2 == 0, img2)
  ax[i , 1].imshow(img1 , cmap = 'gray' , interpolation = 'none')
  ax[i , 1].imshow(img2 , cmap = 'jet', interpolation = 'none', alpha = 0.7)

plt.show()
```

Stvaranje dve NumPy matrice sa prosledjenim dimenzijama.Ove matrice će se koristit za spremanje ulaznih slika (X) i pripadajućih maski (y) u postupku pripreme podataka za obradu ili obuku neuronske mreže.

```python
X = np.zeros((len(train_image), 128, 128, 1), dtype=np.float32)
y = np.zeros((len(train_image), 128, 128, 1), dtype=np.float32)
```

Prolazak kroz listu ulaznih slika i maski. Slika se učitava, pretvara se u NumPy niz, menja se veličina i dodaje treća dimenzija (dubina), a zatim se normaliziju deljenjem sa 255.

```python
for n , img_name in enumerate(train_image):
  img = tf.keras.preprocessing.image.load_img('/content/train/'+img_name , grayscale=True)
  in_img = tf.keras.preprocessing.image.img_to_array(img)
  in_img = skimage.transform.resize(in_img , (128 , 128 , 1) , mode = 'constant' , preserve_range = True)
  X[n] = in_img / 255.0

for n , img_name in enumerate(train_mask):
  img = tf.keras.preprocessing.image.load_img('/content/train/'+img_name , grayscale=True)
  in_img = tf.keras.preprocessing.image.img_to_array(img)
  in_img = skimage.transform.resize(in_img , (128 , 128 , 1) , mode = 'constant' , preserve_range = True)
  y[n] = in_img / 255.0
```

Podela ulaznih podatka matrice X i ciljne vrednosti matrice y na skupove za obuku i validaciju. 10% podataka biti rezervisano za validaciju dok će preostalih 90% biti korišćeno za obuku našeg modela. Nakon izvršenja funkcije, dobijamo četiri nove matrice: X_train, X_valid, y_train i y_valid, koje sadrže podatke za obuku i validaciju. Ova podela nam omogućava da procenimo performanse modela na nepoznatim podacima i da ocenimo njegovu sposobnost generalizacije.

```python
from sklearn.model_selection import train_test_split
X_train , X_valid , y_train , y_valid = train_test_split(X , y , test_size = 0.1 , random_state = 42)
```

Kreiranje modela neuronske mreže funkcijom U_Net. model.compile metoda za konfigurisanje trening postavki modela(optimizer, metrike i gubitak).

```python
img_tensor = tf.keras.layers.Input((128 , 128 , 1) , name = 'img')
model = U_Net(img_tensor)
model.compile(optimizer = tf.keras.optimizers.Adam(),
             loss = 'binary_crossentropy',
             metrics = ['accuracy'])
```

Prikaz sažetka modela koji sadrži informacije o slojevima modela, njihovim dimenzijama i broju parametara.

```python
model.summary()
```

Lista povratnih poziva (callbacks) koje ćemo koristiti tokom obuke modela. Povratni pozivi su objekti koji se izvršavaju u određenim tačkama tokom obuke modela i mogu se koristiti za praćenje metrika, prekidanje obuke, prilagođavanje stope učenja i čuvanje najboljeg modela.

```python
callbacks = [
    tf.keras.callbacks.EarlyStopping(patience=10, verbose=1),
    tf.keras.callbacks.ReduceLROnPlateau(factor=0.1, patience=5, min_lr=0.00001, verbose=1),
    tf.keras.callbacks.ModelCheckpoint('model_nerve.h5', verbose=1, save_best_only=True, save_weights_only=True)
]
```

Ovde treniramo model koristeći metodu fit na skupu za obuku. Postavljamo veličinu grupa (batch size) na 32, što znači da će se 32 uzoraka obraditi istovremeno pre nego što se ažuriraju parametri modela. Takođe, postavljamo broj epoha na 50, što predstavlja broj puta kada će ceo skup za obuku biti prolazak kroz model.

Koristimo povratne pozive (callbacks) koje smo definisali ranije kako bismo pratili metrike tokom obuke i preduzeli odgovarajuće akcije. Povratni pozivi će se aktivirati na određenim tačkama tokom obuke kako bi pružili povratne informacije i preduzeli akcije kao što je zaustavljanje obuke ili smanjivanje stope učenja.

```python
results = model.fit(X_train, y_train, batch_size=32, epochs=50, callbacks=callbacks,\
                    validation_data=(X_valid, y_valid))
```

Kriva učenja (learning curve) model. results.history sadrži vrednosti gubitka tokom svake epohe obuke. val_loss predstavlja gubitak na validacionom skupu podataka tokom svake epohe obuke.  Marker "x" u crvenoj boji označava na grafiku gde je postignut najmanji validacioni gubitak.ž

```python
plt.figure(figsize=(8, 8))
plt.title("Learning curve")
plt.plot(results.history["loss"], label="loss")
plt.plot(results.history["val_loss"], label="val_loss")
plt.plot( np.argmin(results.history["val_loss"]), np.min(results.history["val_loss"]), marker="x", color="r", label="best model")
plt.xlabel("Epochs")
plt.ylabel("log_loss")
plt.legend();
```

Ovde učitavamo težine modela koje su sačuvane u datoteci "model_nerve.h5" koristeći metodu load_weights. Ova metoda postavlja težine modela na vrednosti koje su sačuvane u datoteci, što omogućava da ponovno koristimo prethodno obučeni model.

Nakon učitavanja težina, model će biti spreman za upotrebu i možemo ga koristiti za predviđanja ili evaluaciju na novim podacima.

```python
model.load_weights('model_nerve.h5')
```

Metoda evaluate koristi se da bismo evaluirali performanse modela na validacionom skupu podataka. Metoda evaluate će izračunati gubitak i metrike koje su definisane prilikom kompilacije modela. Rezultat će biti prikazan i možemo ga koristiti da bismo procenili performanse modela na validacionom skupu podataka.

```python
model.evaluate(X_valid, y_valid, verbose=1)
```
